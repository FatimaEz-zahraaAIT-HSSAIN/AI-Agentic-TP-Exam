```python
from dotenv import load_dotenv
import os

# Load the keys from your .env file
load_dotenv(override=True) 

from langchain_groq import ChatGroq
from langchain.messages import SystemMessage, HumanMessage
from IPython.display import display, Markdown

# Get the key from the environment
groq_key = os.getenv("GROQ_API_KEY")

# Initialize ChatGroq with a reliable production model and explicit API key
llm2 = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=groq_key
)

resp2 = llm2.invoke([
    SystemMessage("You are a helpful assistant. The output should be in Markdown"),
    HumanMessage("C'est quoi un Agent AI")
])

# Render the response beautifully
display(Markdown(resp2.content))
```
